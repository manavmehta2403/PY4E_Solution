````python
#Q3
##making exact copy of the count in built function of the python
##here parameters are string, letter to find the in the string and start and end as the index as optional parameter
def count(string,letter,start=None,end=None):
    ##number of appearence of the letter in the string
    appearance = 0
    ##if the start isn't peresent then we assume the starting point is index 0
    if(start is None):
        start = 0
    ## if the end point isn't peresent then we take the last as the last index as one more 
    ##because in slicing last value is not included
    if(end is None):
        end = len(string)
    ##algo from the book with modifiction of the start and end as the index
    for occuerence in string[start:end]:
        if occuerence == letter:
            appearance += 1
    return appearance
##testing the function
string = str(input("Please enter the string: "))
letter = str(input("Please enter the letter(s) "))
print(count(string,letter,0,4))
````

````python
#Q2
##Extracting the values
def extract(string):
    #storing the double value
    double = ""
    ##getting the index :
    start_index = string.find(":")
    
    ##running the loop from : to last index
    for char in string[start_index+1:]:
        ##as it could decimal value thus
        if char == ".":
            double += char
        ## tacking all the digit
        if char.isdigit():
            double += char
    ##returning the decimal
    return float(double)

#prompting user for the file name
file_name = str(input("Please enter the file name with extension: "))
print()
##opening the file
with open("C:\\Users\\manav\\Google Drive (manavmehta2403@hotmail.com)\\Solutions\\PY4E\\"+file_name,"r") as file:
    count = 0
    total = 0
    ##extracting the particular lines
    for lines in file.read().splitlines():
        if lines.startswith("X-DSPAM-Confidence:"):
            ##counting the number of lines
            count += 1
            ##extracting numbers from the file
            total += extract(lines)
    #printing the average
    print("Average spam confidence: ",total/count)
            
````

````python
#Q6
##ist 
num_lit = []
##looping regular
while True:
    ##trying if the inputs are correct
    try:
        ##promt for the user number input
        number = input("Please enter the number: or done for stats: ")

        ##if the input is done then it breaks from the loop
        if (number.casefold() == "done"):
            break
        
        ## if there is number input then it is converted into
        ##float and values is appended in the list
        else:
            number = float(number)
            num_lit.append(number)

    ##just in case when the input is wrong
    except:
        print("\n\tPlease enter number or done.\n")

##printing the stats
print("Max:",max(num_lit))
print("Min:",min(num_lit))
print("Sum:",sum(num_lit))
print("Avg:",sum(num_lit)/len(num_lit))
````

````python
#Q5
##prompt the user for the file name
fhand = input("Please enter the file name. ")

fhand = open("C:\\Users\\manav\\Google Drive (manavmehta2403@hotmail.com)\\Solutions\\PY4E\\"+fhand,"r")

##counter dictionary for holding counts
counter = {}
##taking the domain name
domain = []


for line in fhand.read().splitlines(): 
    words = line.split()
    ##lines start from
    if len(words) == 0 or words[0] != 'From': 
        continue
    ##finding the index of @
    start_index = words[1].find("@")
    ##scraping the result from @
    domain.append(words[1][start_index+1:])

##using the get to count
for portal in domain:
    counter[portal] = counter.get(portal,0) + 1

print(counter)
````
